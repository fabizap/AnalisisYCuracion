# MATPLOTLIB


```python
import matplotlib.pyplot as plt
```

```python
a = [1,2,3,5,7,8]
b = [9,8,5,4,6,2]
c = [0,4,6,9]
d = [5,3,6,8]
```

configura caracteristicas del grafico

```python
plt.plot (a, b, color='orange', linewidth=3, label='nortworth')
plt.plot (c, d, color='yellow', linewidth=3, label='surthwis')

plt.title ('nombre graf')
plt.xlabel('meses')
plt.ylabel('ganancias')
plt.legend()
plt.show ()
```
